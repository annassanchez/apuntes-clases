# ¿Qué es la terminal?

- **GUI** (*Graphical User Interface*): interacción usuario-máquina a través de gráficos, iconos, imágenes, textos.


- **CLI** (*Command Line Interface*): interacción usuario-máquina a través de comandos de texto, a través de la terminal.

> **Sinónimos de CLI**:
>- Shell
>- Console
>- Terminal

**Terminal** o **Consola**, se define como cualquier dispositivo electrónico que forma parte del Hardware de un ordenador, y que tiene la funcionalidad básica de introducir o visualizar los datos que se encuentran dentro de un ordenador o en un determinado sistema informático. Por ello, podemos definir que su finalidad básica es la de transformar estos datos en información. 

De esta manera podemos distinguir entre dos tipos de terminales:

- macOS 🍎 y Linux 🐧 --> **Unix**



- Windows 🪟 --> **Git Bash** 


> Aunque los dos funcionan de manera similar, puede haber algunos comandos que son un poco diferentes en Windows.   
> [Aquí](https://enexdi.sciencesconf.org/data/pages/windows_vs_mac_commands_1.pdf) hay una tabla de equivalencias de comandos.


# Conceptos básicos

- Algunos comandos **NECESITAN** opciones y/o argumentos!


- Usando la tecla`<TAB>` autocompletamos el código

# Comandos básicos

Algunos comandos que pueden ser útiles cuando empezamos a trabajar con Bash:

- `man <comando>` -> muestra el manual de ese comando. ⚠️ Control+Z para salir.


- `clear` -> elimina todos los comandos previos y los outputs de la terminal.


- `history` -> nos muestra todos los comandos que hemos escrito en la terminal. 

## Operaciones de directorios

- `pwd` - > nos muestra el directorio donde estamos. 


- `mkdir <nueva_carpeta>` -> creamos una nueva carpeta llamada "nueva_carpeta".


- `cd <nueva_carpeta>` -> cambiamos la carpeta donde estamos, nos movemos a la carpeta "nueva_carpeta".


- `cd ..`: vamos una carpeta para atrás.


- `cd` -> vamos a la raíz de nuestro ordenador. 


- `ls` -> lista todos los ficheros y carpetas que tenemos donde estamos.  list directory contents. El comando ls nos permite ver rápidamente todos los archivos dentro del directorio especificado.


    - ls -a -> lista todos los ficheros dentro de la carpeta, incluidos los ocultos. 
    
    - ls -l -> lista con detalles.
    
    - ls -R -> lista con recursividad.

## Operaciones con ficheros


**Crear, Editar y Explorar**

- `echo <hola>` -> imprime por pantalla la palabra "hola".


- `touch <fichero1>` -> creamos un fichero que se llama "fichero1".


- `echo hola >> <fichero1>`: apendea la palabra "hola" al "fichero1". 



- `cat <fichero1>` -> vemos el contenido de "fichero1".



- `echo adiós > <fichero1>`: inserta la palabra "adiós" al "fichero1". 🚨 Cuidado porque elimina todo o que hubiera previamente en ese fichero.




- `head <fichero1>` -> vemos las 10 primeras lineas del "fichero1".


    - `head -n 3 <fichero1>` -> vemos las 3 primeras lineas del "fichero1".
    

- `tail <fichero1>` -> vemos las 10 últimas lineas del "fichero1".


    - `tail -n 3 <fichero1>` -> vemos las 3 últimas 3 lineas del "fichero1".



- `nano <fichero1>` -> edita el contenido del "fichero1".



- `wc <fichero1>` -> cuenta lineas/palabras/caracteres del "fichero1".



- `cp <fichero1> <data>` -> copia "fichero1" a la carpeta data.  ⚠️ NOTA: lo que quiero copiar seguido de dónde quiero copiarlo.


- `mv <fichero1> <data>` -> mueve "fichero1" a la carpeta "data".  ⚠️ NOTE: Es como cortar un fichero! 


- `rm <fichero1>` -> elimina **FICHEROS**. Elimina el "fichero1". 


- `rm -r <carpeta1>` - elimina **CARPETAS**. Elimina la "carpeta1".


## Buscar ficheros

> `*` : comodín  
> `|` : ejecutar el comando sobre el resultado de otro comando (cat archivo | grep patrón)



- `find <name*>` -> busca los ficheros que empiezan com "name"


    - `find </dir/> -name '<name*>'` -> busca los ficheros que empiezan com "name" en la carpeta "dir"
    

- `grep <patrón> <fichero1>` -> busca el "patrón2 en el "fichero1"


- `grep -r <patrón> <fichero1>` -> búsqueda recursiva


- `grep -o <patrón> <fichero1>` -> muestra solo la parte donde encuentra la búsqueda. 


## Gestión de procesos

- `top`-> muestra los procesos en tiempo real. ⚠️ Control+Z para salir.


- `ctrl + R`: para buscar comandos previos.
