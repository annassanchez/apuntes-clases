```python
def formatear_fecha(mes, diccionario_conversion = diccionario):
    for k, value in diccionario_conversion.items():
        if k == mes:
            return value
```

```python
def formatear_fecha(mes, diccionario_conversion = diccionario):
    return diccionario_conversion[mes]
```
