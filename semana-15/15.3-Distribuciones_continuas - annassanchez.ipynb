# Distribuciones continuas

Las variables aleatorias de tipo continuo son aquellas que tratan de medir cantidades que pueden tomar cualquier valor entre ciertos límites. Además, se definen como aquellas variables cuya función de distribución es absolutamente continua. 

Recordamos que tenemos:

- Distribución normal


- Distribución uniforme


- Distribución exponencial

# Distribución normal

La distribución normal es la distribución de probabilidad más importante en estadística, porque se ajusta a muchos fenómenos naturales y sociales :

 - las alturas
 
 - la presión arterial
 
 - las puntuaciones del coeficiente intelectual
 
También se conoce como la campana de Gauss. 

Se distribuye de la siguiente manera: 

$Y \sim N(\mu, \sigma)$

Donde:
 - $\mu$ será la media
 
 - $\sigma$ será la desviación estándar


Usando el método de scipy recibe dos parámetros: 

- `loc` que corresponde con la media


- `scale` que corresponde con la desviación estándar

```python
# vamos a definir dos variables que representan la estatura media y la desviación de los alumnos de una clase

altura_media = 170
desviacion = 20
```

Intervalos de confianza

($\mu - \sigma, \mu + \sigma)$

($\mu - 2\sigma, \mu + 2\sigma)$

($\mu - 3\sigma, \mu + 3\sigma)$

```python
# inicializamos nuestra distribución normal usando el método "norm"

my_normal = norm(loc=altura_media, scale=desviacion)
```

```python
# si sacamos la media de la distribución que acabamos de crear

my_normal.mean()
```

```python
# si sacamos la desviación estándar de la distribución que acabamos de crear

my_normal.std()
```

Vamos a usar los métodos que nos permiten "jugar" con nuestra distribución:

```python
# simulamos 10 eventos normales usando el rvs

normal_30 = my_normal.rvs(size=30)
normal_30
```

```python
# veamos como se ve

sns.histplot(normal_30, kde = True, color= "grey");
```

```python
# y que pasa si aumentamos el tamaño muestral? 

normal_300 = my_normal.rvs(size=300)
normal_300[:50]
```

```python
sns.histplot(normal_300,  kde = True, color="grey");
```

```python
# y si lo aumentamos más?

normal_3000 = my_normal.rvs(size=3000)
normal_3000[:50]
```

```python
sns.histplot(normal_3000, kde = True, color = "grey");
```

Si ampliamos la muestra, vemos que el comportamiento de ls alturas sí que son normales -- por tanto, cuanto más datos mejor.

Vemos que, a mayor tamaño muestral mejor se ajusta nuestros datos a la distribución que estamos definiendo. 

```python
# generamos notas de 100 alturas aleatoriamente con intervalos continuos

nuevas_personas= np.linspace(120, 220, 100)
nuevas_personas[:50]
```

```python
# De los valores generados anteriormente, vamos a calcular la probabilidad de que una persona pueda medir cada uno de los valores definidos
# es como el pmf, pero tiene que ser para las funciones continuas.
prob_nuevas = my_normal.pdf(nuevas_personas)
prob_nuevas[:50]
```

```python
# veamoslo graficamente

plt.plot(nuevas_personas,prob_nuevas);
```

```python
# ¿Cuál es la probabilidad de que una persona mida menos de 140?
menos_140 = my_normal.cdf(140)
menos_140
#mayor de 140 sería 1 - menos_140
```

```python
# y de que mida menos de 160?

menos_160 = my_normal.cdf(160)
menos_160

#mayor de 160 sería 1 - menos_160
```

Veamos todo esto graficamente: 

```python
# creamos un array para definir las zonas que vamos a querer remarcar en otro color
fill_x = np.linspace(120, 160, 100)

#utilizamos el método fill_between para decirle a matplotlib que nos rellene una zona del plot
plt.fill_between(fill_x, my_normal.pdf(fill_x),color='r')

# incluimos un texto en el gráfico
plt.text(150, 0.003, round(menos_160,2), size=15)

# finalmente lo ploteamos
plt.plot(nuevas_personas,prob_nuevas);
```

Calculamos la probabilidad de que una persona este dentro de un intervalo --> ($\mu - \sigma, \mu + \sigma)$

$$P(150 < X < 190)$$

```python
intervalo = my_normal.cdf(altura_media + desviacion) - my_normal.cdf(altura_media - desviacion)
intervalo
```

```python
# lo vemos graficamente

fill_x2 = np.linspace(150, 190, 100)

plt.fill_between(fill_x2, my_normal.pdf(fill_x2),color='r')
plt.text(165, 0.01, intervalo, size=15)

plt.plot(nuevas_personas,prob_nuevas);
```

# Distribución uniforme


La distribución uniforme toma cualquier valor situado entre a y b siendo esta la misma 


Se distribuye de la siguiente manera: 

$X \sim U(a, b)$

Donde:
 - $a $ valor minimo  
 
 - $b $ valor maximo
 
Por ejemplo: $X \sim U(0, 1)$



Entendamoslo mejor con algunos ejemplos: 

$P(x \geq 0)=1$

$P(x \leq 1)= 1$

$P(x \geq 0.7)=0.3$

$P(x \leq 0.4)=0.4$

$P(0.1 \leq x \leq 0.4)=0.4 - 0.1=0.3$

```python
# iniciamos nuestra distribución uniforme entre 0 y 1

my_uniform = uniform(0, 1)
```

```python
# generamos valores uniformes

uniforme_30 = my_uniform.rvs(size=30)
uniforme_30
```

```python
# lo vemos visualmente

sns.histplot(uniforme_30, color = "gray", bins = 5);
```

```python
# sigamos aumentando nuestro tamaño muestral

uniforme_300 = my_uniform.rvs(size=300)
uniforme_300[:50]
```

```python
sns.histplot(uniforme_300, color = "gray", bins = 15);
```

```python
# y ahora un poco más grande

uniforme_3000 = my_uniform.rvs(size=3000)
uniforme_3000[:50]
```

```python
sns.histplot(uniforme_3000, color = "gray", bins = 20);
```

```python
# y ahora un poco más grande

uniforme_300000 = my_uniform.rvs(size=30000000)
uniforme_300000[:50]
```

```python
sns.histplot(uniforme_300000, color = "gray", bins = 20);
```

De nuevo vemos como a medida que aumentamos el tamaño muestral, los datos se ajustan mejor a la distribución deseada

```python
# ¿qué probabilidad hay de que sea 0.4?

my_uniform.pdf(0.4)
```

```python
# independiente del valor, si es de 0 a 1 será 1 porque es uniforme

my_uniform.pdf(0.8)
```

```python
# ¿y de qué sea 4?

my_uniform.pdf(4)
```

Este valor es 0 porque el 4 esta fuera de nuestro intervalo, ya que nuestra normal va de 0-1

```python
# generemos 10 datos aleatorios entre el 0-1 usando métodos de numpy

nuevos_uniform = np.random.rand(10)
nuevos_uniform
```

```python
# y calculemos la probabilidad de ocurrencia de cada uno de los números creados anteriormente

for k in nuevos_uniform:
    ocurrencia = my_uniform.pdf(k )
    print(f"La probabilidad de {k} es: {ocurrencia}")
```

Que recordamos que es lo mismo que: 

```python
my_uniform.pdf(nuevos_uniform)
```

# Distribución exponencial 

La distribución exponencial describe el tiempo transcurrido entre dos sucesos consecutivos de Poisson o el tiempo de espera hasta que ocurre un suceso de Poisson.


Se distribuye de la siguiente manera: 

$X \sim Exp(\lambda)$

Donde:
 - $\lambda$: tiempo medio transcurrido entre dos sucesos de Poisson. 

Algunos ejemplos de distribucion exponencial son:

- el tiempo de espera hasta que un calvo aparezca por ih 
- tiempo que tarda en mutar el covid a una nueva variante 



En media entra un calvo cada 9 días. Vamos a calcular la probabilidad de que entre un calvo en dos días. 

```python
# iniciamos la distribución exponencial

my_e = expon(scale=9)
```

```python
# generamos 3000 eventos donde simulamos que entran calvos

expo_3000 = my_e.rvs(size=3000)
expo_3000[:50]
```

se mide el tiempo de espera entre dos eventos de poisson -- en este caso mi fenómeno es que aparezca un calvo

```python
# veamoslo visualmente

sns.histplot(expo_3000, kde = True, color = "gray");
```

```python
# probabilidad de que en tres dias entre un calvo

my_e.pdf(3)
```

```python
# definimos números pseudoaleatorios simulando el número de dias en media que tarda en entrar un calvo

dias_calvo = np.linspace(0, 100, 100, dtype = int)
dias_calvo
```

```python
# visualmente, la probabilidad de que aparezca un calvo aumenta

prob_calvos = my_e.pdf(dias_calvo)
plt.plot(dias_calvo,prob_calvos);
```

La probabilidad de que pasen 20 días y entre un calvo es de aproximadamente 0.018

```python
# ¿Cuál es la probabilidad de que en 2 días entre un calvo?

my_e.cdf(2)
```

```python
# ¿y en 10?

my_e.cdf(10)
```

```python
# ¿y en 20? acumuladamenteg

my_e.cdf(20)
```
