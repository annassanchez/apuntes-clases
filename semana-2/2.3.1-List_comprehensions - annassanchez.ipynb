# ¿Qué son? 

Son una forma de crear listas de una manera elegante simplificando (en una línea de código) el código al máximo. Es básicamente crear una lista en función de otra lista.  

La ventaja de utilizar comprensiones de lista:

- Son relativamente más rápidas que los bucles for.


- La sintaxis de las comprensiones de lista es más fácil de leer y entender.

```python
%%time
cuadrado = []
for i in range(5001):
    cuadrado.append(i ** 2)
cuadrado
```

```python
%%time
cuadrado2 = [i ** 2 for i in range(5001)] ## el condicional va después de la operación
```

![list-comp.png](https://github.com/Ironhack-Data-Madrid-PartTime-Oct22/imagenes/blob/master/semana-2/list_compre.png?raw=true)

```python
palabras = ['data', 'science', 'machine', 'learning', 'math', 'python', 'code']
```

Queremos crearnos una lista donde tengamos las longitudes de las palabras


```python
%%time
long = []

for palabra in palabras:
    long.append(len(palabra))
```

```python
# list comprehension
%time
long2 = [len(palabra) for palabra in palabras]
```

```python
## lista palabras
## long > 4
## cambiar e por A
nuevas_palabras = []
for palabra in palabras:
    if len(palabra) > 4:
        #palabra.append(nuevas_palabras)
        palabra = palabra.replace('e', 'A')
        nuevas_palabras.append(palabra)
        #print(palabra.replace('e', 'A'))
```

```python
## lista palabras
## long > 4
## cambiar e por A
nuevas_palabras = []
for palabra in palabras:
    if len(palabra) > 4:
        nuevas_palabras.append(palabra.replace('e', 'A'))
```

```python
nuevas_palabras2 = [palabra.replace('e', 'A') for palabra in palabras if len(palabra) > 4]
nuevas_palabras2
```

# *Dict comprehension*

![image.png](https://github.com/Ironhack-Data-Madrid-PartTime-Oct22/imagenes/blob/master/semana-2/dict_comp.png?raw=true)

Queremos crear un diccionario donde:
   
- Las `keys` cada número que tenemos en la lista


- Los `values` las veces que aparece cada número en nuestra lista

```python
# tradicional
%time
num_list = [1,2,31,2,3,1,5,3,4,2,3,4,5,6]
mi_diccionario = {}
for num in num_list:
    mi_diccionario[num] = num_list.count(num)
mi_diccionario
```

```python
%time
mi_diccionario2 = {numero: num_list.count(numero) for numero in num_list}
mi_diccionario2
```
