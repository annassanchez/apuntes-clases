**Utilizando la columna de `reservation_status_date`:**

- Cread dos columnas nuevas que correspondan con  el mes y el año. 

- Una vez que las hayais creado eliminad la columna `reservation_status_date`

- ¿Qué mes es en el que tenemos más reservas? ¿Y el año?

```python
def separar_fechas(fecha):
    lista_fecha = fecha.split("-")
    return lista_fecha[0], lista_fecha[1]
```

```python
df[["year", "month"]] = df.apply(lambda x: separar_fechas(x["reservation_status_date"]), axis = 1, result_type= "expand")
```

```python
df.drop(["reservation_status_date"], axis = 1, inplace = True)
```

**Ahora no nos interesan los meses en números y queremos tenerlos con letra, es decir, enero, febrero, marzo, etc... Cambiad los valores de la columna "month" para tenerlos en el formato que queremos.**

```python
meses_numericos = sorted(list(df["month"].unique()))
meses_letras = list(calendar.month_name)[1:]
```

```python
diccionario = dict(zip(meses_numericos, meses_letras))
diccionario
```

```python
def formatear_mes(mes, diccionario_conversion = diccionario):
    for k, v in diccionario.items():
        if k == mes:
            return v
    
```

```python
df["month"] = df["month"].apply(formatear_mes)
```
