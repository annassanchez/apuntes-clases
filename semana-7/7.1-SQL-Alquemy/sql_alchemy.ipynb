# CASE
` (CASE 
    
    WHEN condition1 THEN result1
    
    WHEN condition2 THEN result2
    
    WHEN condition3 THEN result3
    
    ELSE condition4

END) ` AS


```python
query_case = """
    SELECT (
        CASE
            WHEN experiencia = '0-9y' THEN 'cliente_joven'
            WHEN experiencia = '10-19y' OR experiencia = '20-29y'  THEN 'cliente_medio'
            ELSE 'cliente_senior'
    END) AS tipo_cliente, experiencia, idcliente, genero, raza
    FROM clientes;

"""
```

```python
df_case = pd.read_sql(query_case, engine)
df_case
```

## ejercicio
 case con múltiples joins

```python
query_case2 = """
    SELECT (
        CASE
            WHEN score < 0.3 THEN 'cliente_de_riesgo'
            WHEN score > 0.3 AND score < 0.8  THEN 'regulinchi'
            ELSE 'cliente_suuuper'
    END) AS tipo_cliente_score, score, genero, educacion, genero, idvehiculos
    FROM seguro
    INNER JOIN clientes
    ON seguro.clientes_idcliente = clientes.idcliente
    INNER JOIN vehiculos
    ON clientes.idcliente = vehiculos.clientes_idcliente
    ;

"""
```

```python
df_case2 = pd.read_sql(query_case2, engine)
df_case2.head()
```

```python
df_case2['tipo_cliente_score'].unique()
```

```python
query_sub3 = """
    SELECT idcliente, genero, estado_civil, clase, codigo_postal
    FROM clientes
    WHERE educacion = "university"
    and idcliente in (  SELECT idcliente
                        FROM clientes
                        WHERE hijos = 1
                        AND raza = "minority"
                        AND experiencia = "30y+")
"""
```

```python
pd.read_sql(query_sub3, engine)
```

```python
## otra forma de hacer la subquery
query_sub3 = """
    SELECT idcliente, genero, estado_civil, clase, codigo_postal
    FROM (
        SELECT *
        FROM clientes
        WHERE hijos = 1
        AND raza = "minority"
        AND experiencia = "30y+"
        AND educacion = 'university'
        ) as tabla1
"""
```

```python
pd.read_sql(query_sub3, engine)
```
